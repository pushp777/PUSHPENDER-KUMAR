```python
#import all the required libraries
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

## Data Scraping from Amazon

### Scraping daily use electronic items.

```python
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.amazon.in')

search=driver.find_element_by_id('twotabsearchtextbox')  
search.send_keys('daily use electronic items')
button=driver.find_element_by_id('nav-search-submit-button')
button.click()
```

```python
start_page = 0
end_page = 5
urls = []
for page in range(start_page,end_page+1):
    try:
        page_urls = driver.find_elements_by_xpath("//h2/a")
        
        # appending all the urls on current page to urls list
        for url in page_urls:
            url = url.get_attribute('href')     
            if url[0:4]=='http':                
                urls.append(url)                
        print("Product urls of page {} has been scraped.".format(page+1))
        
        # Moving to next page
        nxt_button = driver.find_element_by_xpath('//li[@class="a-last"]/a')      
        if nxt_button.text == 'Nextâ†’':                                            
            nxt_button.click()                                                    
            time.sleep(8)                                                         
          
        elif driver.find_element_by_xpath('//li[@class="a-disabled"]').text == 'Nextâ†’':    
            print("No new pages exist. Breaking the loop")  
            break
            
    except StaleElementReferenceException as e:            
        print("Stale Exception")
        next_page = nxt_button.get_attribute('href')        
        driver.get(next_page)   
print(len(urls))        
```

```python
Review=[]
short_review=[]
Rating=[]
for u in urls:
    driver.get(u)
    
    rrr=driver.find_elements_by_xpath('//div[@class="a-row"]/a')
    for i in rrr:
        if len(i.get_attribute('title'))==18:
            t=i.get_attribute('title')
            Rating.append(t)
        
        # scraping for review
    rr=driver.find_elements_by_xpath('//div[@id="cm-cr-dp-review-list"]/div/div/div/div[4]')
    for i in rr[0:8]:
        t=i.text
        Review.append(t)

    
        # scraping for short review
    sr=driver.find_elements_by_xpath('//a[@data-hook="review-title"]/span')
    for i in sr[0:8]:
        t=i.text
        short_review.append(t)
print(len(Rating),len(Review),len(short_review)) 
```

```python
# create data set
df_mix=pd.DataFrame()
df_mix["Review"]=Review[0:2581]
df_mix['ShortReview']=short_review[0:2581]
df_mix['Rating']=Rating[0:2581]
df_mix.to_csv('mix_data.csv')
df_mix
```
